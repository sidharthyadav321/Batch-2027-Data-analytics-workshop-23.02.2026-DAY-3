```python
import statistics
import numpy as np

numbers = [10, 20, 30, 40, 50, 20]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)
variance = statistics.variance(numbers)
std_dev = statistics.stdev(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

# Sum and Mean
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))

# Max and Min
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))

# Reshape (2x3 matrix)
reshaped = arr.reshape(2, 3)
print("Reshaped Array:\n", reshaped)

# Element-wise multiplication
result = arr * 2
print("Element-wise Multiplication:\n", result)
```

```python
import numpy as np

# Create matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Addition
print("Addition:\n", A + B)

# Subtraction
print("Subtraction:\n", A - B)

# Matrix Multiplication
print("Multiplication:\n", np.dot(A, B))

# Transpose
print("Transpose of A:\n", A.T)
```

```python
import numpy as np

random_arr = np.random.randint(1, 100, 10)

print("Array:", random_arr)
print("Mean:", np.mean(random_arr))
print("Standard Deviation:", np.std(random_arr))
print("Minimum:", np.min(random_arr))
print("Maximum:", np.max(random_arr))
```

```python
import numpy as np
from scipy import stats

data = [10, 12, 9, 11, 10, 13, 12]

# Mean using scipy.stats
mean = stats.tmean(data)
print("Mean:", mean)

# One-sample t-test (compare with population mean = 10)
t_stat, p_value = stats.ttest_1samp(data, 10)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```
