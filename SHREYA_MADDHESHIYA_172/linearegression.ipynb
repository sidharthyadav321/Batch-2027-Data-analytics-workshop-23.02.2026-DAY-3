```python
import statistics
import numpy as np

numbers = [10, 20, 30, 40, 50, 20]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)
variance = statistics.variance(numbers)
std_dev = statistics.stdev(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import numpy as np
from scipy import stats

data = [10, 12, 9, 11, 10, 13, 12]

# Mean using scipy.stats
mean = stats.tmean(data)
print("Mean:", mean)

# One-sample t-test (compare with population mean = 10)
t_stat, p_value = stats.ttest_1samp(data, 10)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
from scipy import integrate
import numpy as np

# Function: f(x) = x^2
def func(x):
    return x**2

result, error = integrate.quad(func, 0, 3)

print("Integration Result:", result)
print("Estimated Error:", error)
```

```python
from scipy import optimize

# Function: f(x) = x^2 + 4x + 4
def func(x):
    return x**2 + 4*x + 4

result = optimize.minimize(func, x0=0)

print("Minimum Value:", result.fun)
print("Minimum at x =", result.x)
```

```python
# WAP to implement Simple Linear Regression

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

n = len(x)

sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum([x[i]*y[i] for i in range(n)])
sum_x2 = sum([x[i]**2 for i in range(n)])

# Calculate slope (b1)
b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)

# Calculate intercept (b0)
b0 = (sum_y - b1 * sum_x) / n

print("Slope (b1):", b1)
print("Intercept (b0):", b0)

# Prediction
x_new = 6
y_pred = b0 + b1 * x_new
print("Predicted value for x =", x_new, "is", y_pred)
```
